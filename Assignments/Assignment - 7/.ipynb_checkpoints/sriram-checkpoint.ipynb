### Application of PCA on real dataset

**(4a) Load the human face data set from sklearn (fetch_lfw_people).**

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
```

**(4b) Apply PCA on the faces data (sklearn can be used). Take the top 150 features from PCA. Reconstruct the images with the these 150 features and visualise them. Give your observations.**

```python
# Write your code here
data5=faces.data
from sklearn.decomposition import PCA
p=PCA(n_components=150)
p.fit(data5)
face_tr = p.transform(data5)
face_final = p.inverse_transform(face_tr)
```

```python
N, h, w = faces.images.shape
images_show = list(range(10))
image_actual = np.reshape(data5, (N, h, w))[images_show]
image_reconstructed = np.reshape(face_final, (N, h, w))[images_show]

fig, axes = plt.subplots(10, 2, figsize=(20,20))
for idx, row in enumerate(axes):
    true_img = image_actual[idx]
    recon_img = image_reconstructed[idx]
    row[0].imshow(true_img, cmap='gray')
    row[0].set_title('Actual')
    row[1].imshow(recon_img, cmap='gray')
    row[1].set_title('Reconstructed')
fig.tight_layout()
plt.show()
```

<b> The reconstructed images are a blurred version of the original image. This is due to the fact that some information is lost when we retain only the first 150 PCs.<b>

### 5) Cross validation to select parameters for Ridge Regression 

$$\hat{\theta} = argmin_{\theta}((y - X\theta)^\top(y - X\theta) + \lambda \theta^\top \theta)$$


Perform ridge rigression using gradient descent on the data 4 provided. (You can use sklearn)

Try to fit the data with polynomial of degree 3 and 

- For $\lambda$ choose from = [10,100,1000]
- choose learning rate (here it is $\alpha$)= [ 0.01,0.001,0.0001]
- max_iters = 1000

Choose the best combination of $\lambda$ and $\alpha$ using 5-fold cross validation on the dataset provided and report the same.


```python
# Write your code here
data4 = np.loadtxt('data_X.txt')
data4=pd.DataFrame(data4)
x=data4.drop(axis=1,columns=[13])
y=data4[13]
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
lamda=[10,100,1000]
p=PolynomialFeatures(3)
xtrain=p.fit_transform(x)
score=[]
for i in lamda:
    r=RidgeCV(alphas=[i],cv=5)
    r.fit(xtrain,y)
    score.append(r.score(xtrain,y))
print(score)
```

<b>Hence we will use lamda of 100 for this regularization<b>
