Write a function for doing one hot encoding for all categorical features

Hint: Use pandas.get_dummies

```python

def onehot_encoding(df,columns):
    '''
    Takes the dataframe 
    columns which are corresponding to categorical features
    '''
    for c in columns:
        onehot = pd.get_dummies(df[c])
        df[onehot.columns] = onehot
    df.drop(columns,axis=1,inplace=True)
    return df
columns = ['Street','Condition1','Condition2','CentralAir','HeatingQC','LotShape','LandContour']
onehot_encoding(df,columns=columns)
```

```python
import numpy as np
from matplotlib import pyplot as plt

alphas = np.logspace(-1, 3, 100)

plt.figure(figsize=(5, 3))

for Model in [linear_model.Lasso]:
    scores = [1-1*cross_val_score(Model(alpha), X_train, y_train,scoring='r2', cv=5).mean()
            for alpha in alphas]
    plt.plot(alphas, scores,  label=Model.__name__)

plt.legend(loc='left')
plt.xlabel('alpha')
plt.ylabel('cross validation score')
plt.tight_layout()
plt.show()
```
