## 1. SVM (Linearly Seperable Data)

### You are given a two dimensional dataset (SVM_Q1) with 4 classes and you have to use SVM for classification.

### Import the required libraries 

```python
# Imports library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### The code for loading data is given below.

```python
# load data (do not change it)
import numpy as np
X_train, X_test, y_train, y_test = np.load('./SVM_Q1.npy', allow_pickle=True)
```

### 1.a) Write code to visualize the training data

```python
# Generate scatter plot for training data 
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,s=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('SVM_Q1 Data: Each color represents one class')
plt.show();
```

### 1.b) Write the code for SVM classifier (you can write code from scratch or use sklrean)

```python
# SVM classifier code here
from sklearn.svm import SVC
clf = SVC(gamma='auto')
clf = clf.fit(X_train,y_train)
y_test_pred = clf.predict(X_test)
```

### 1.c) Generate the confusion matrix for the classifer on the test data

```python
# Generate confusion matrix
from sklearn.metrics import confusion_matrix
print('Confusion Matrix:\n',confusion_matrix(y_test,y_test_pred))
```

### 1.d) Visualize the support vectors and decision boundary

```python
# Code here

def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy
def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contour(xx, yy, Z, **params)
    return out

fig, ax = plt.subplots()
# title for the plots
title = ('Decision surface of linear SVC')
# Set-up grid for plotting.
X0, X1 = X_train[:, 0], X_train[:, 1]

xx, yy = make_meshgrid(X0, X1)
plot_contours(ax, clf, xx, yy, alpha=0.8)
ax.scatter(X0, X1, c=y_train, s=1.5)
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=1.5,
           linewidth=1, facecolors='none', edgecolors='r',label='Support Vectors')
ax.set_ylabel('x1')
ax.set_xlabel('x2')
ax.legend()
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(title)
plt.show()
```

### 1.e) Report the accuracy obtained 

```python
# Code for Accuracy

print('Prediction Accuracy obtained on test set is: %.3f'%clf.score(X_test,y_test))
```
